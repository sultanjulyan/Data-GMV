```python
import pandas as pd

dataset = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv')
```

```python
print("Lima Data Teratas: ")
print(dataset.head())
```

```python
## Exploratory Data Analyst
## Data Mining, Data Cleansing, Analyst

print("Ukuran dataset: %d baris dan %d kolom \n" % dataset.shape)
```

```python
## Order Month
import datetime

dataset['month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d").strftime('%Y-%m'))
```

```python
dataset.head()
```

```python
# GMV Gross Merchandise Value <- Indikator penjualan
# GMV = item_price * qty <- Mendapatkan value penjualan

dataset['gmv'] = dataset['item_price']*dataset['quantity']
```

```python
dataset.head()
```

```python
# Summary GMV based on bulan

monthly_gmv = dataset.groupby('month')['gmv'].sum().reset_index()
monthly_gmv.head()
```

```python
# First Plot

import matplotlib.pyplot as plt

plt.plot(monthly_gmv['month'], monthly_gmv['gmv'])
plt.show()
```

```python
# Melalui library pandas

dataset.groupby(['month'])['gmv'].sum().plot()
```

```python
plt.figure(figsize=(15,5))
dataset.groupby(['month'])['gmv'].sum().plot()

plt.show()
```

```python
# Title dan axis label

plt.figure(figsize=(15,5))
dataset.groupby(['month'])['gmv'].sum().plot()
plt.title("Hasil penjualan GMV 2019")
plt.xlabel("Order Month")
plt.ylabel("Total GMV")

plt.show()
```

```python
# Custom Graph

plt.figure(figsize=(15,5))
dataset.groupby(['month'])['gmv'].sum().plot()
plt.title("Hasil penjualan GMV 2019", loc="center", pad=40, fontsize=20, color='blue')
plt.xlabel("Order Month", fontsize=15)
plt.ylabel("Total GMV", fontsize=15)

plt.show()
```
